1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

json.loads(response)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

import json
data = json.loads(response)
print("標題:")
for d in data:
    print(d['title'])
```

```python
print("貼文時間:")
for e in data:
    print(e['createdAt'])
```

```python
print("留言人數:")
for f in data:
    print(f["commentCount"])
```

```python
print("按讚人數")
for g in data:
    print(g["likeCount"])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

sum = 0.0
for g in data:
    print(g["likeCount"])
    sum += int(g)
    
print (sum / len(g["likeCount"]))
```
