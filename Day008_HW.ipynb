### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)
# 想要爬資料的目標網址
response = r.text
# 模擬發送請求的動作

#r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')
print(response)

print(type(response))
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

print(type(response))

# print(response['message'])
print(response[:100])

###str 字串型式
```
